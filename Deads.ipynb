### Cancer seems to be increasing through the years, so lets analyse this:
##### Lets import some data about Cancer disease... 

```python
cancer = pd.read_csv('total-cancer-deaths-by-type.csv')
cancerArg = cancer.loc[cancer['Entity']=='Argentina']
cancerArg.drop(columns=['Entity','Code'])
cancerArg = cancerArg.set_index('Year')
cancerArg = cancerArg.drop(columns =['Entity','Code'])
```

```python
cancerArg
```

#### We need to know how many people died because of cancer each year. We have the percentege of people and the population each year

```python
cancerList = causesArg['Cancers (deaths)'].values.tolist()
```

```python
popCancer = [] 
for i in range(len(cancerArg)):
    product = cancerList[i]*popList[i][0]/100
    popCancer.append(product)
```

```python
popCancer #ammount of people how died every year because of cancer since 1990 to 2017
```

### with popCancer I can estimate the percentege of people who has an specific cancer

```python
for year in range(1990,2018):
    pos =+ 1
    cancerArg.loc[year] = 100*(cancerArg.loc[year]/popCancer[pos+1])
```

```python
cancerArg
```
